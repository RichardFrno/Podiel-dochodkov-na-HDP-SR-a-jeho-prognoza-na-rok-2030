```python
# Importing libraries
import pandas as pd
import plotly.graph_objects as go
```

```python
pwd
```

```python
# Importing data
data = pd.read_excel('AgePyramidData.xlsx')
display(data)
```

```python
y_age = data['Age']
x_M = data['2020-Male']
x_F = data['2020-Female'] * -1
```

```python
# Creating instance of the figure
fig = go.Figure()
  
# Adding Male data to the figure
fig.add_trace(go.Bar(y= y_age, x = x_M, 
                     name = 'Mu≈æi', 
                     orientation = 'h'))
  
# Adding Female data to the figure
fig.add_trace(go.Bar(y = y_age, x = x_F,
                     name = '≈Ωeny', orientation = 'h'))
  
# Updating the layout for our graph
fig.update_layout(title = 'Vekov√° pyram√≠da popul√°cie SR v roku 2020',
                 barmode = 'relative',
                 bargap = 0.0, bargroupgap = 0,
                 xaxis = dict(tickvals = [-60000, -50000, -40000, -30000, -20000, -10000,
                                          0, 10000, 20000, 30000, 40000, 50000, 60000],
                                
                              ticktext = ['60K', '50K', '40K', '30K','20K', '10K', '0', 
                                          '10K', '20K', '30K', '40K', '50K', '60K'],
                                
                              title = 'Poƒçet obyvateƒæov v tis√≠coch',
                              )
                 )
  
fig.show()
```
