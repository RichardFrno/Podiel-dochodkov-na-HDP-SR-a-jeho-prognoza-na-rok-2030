```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_excel('CumulativePensionsExpanditure.xlsx', index_col='Rok')
```

```python
data.head()
```

```python
title='Kumulatívne výdavky Sociálnej poisťovne na dôchodky v období 2009-2020'
ylabel='v mld. €'
xlabel='Rok'

ax = data.plot.bar(figsize=(12,6),title=title, grid=True, stacked=True, color={"Starobné": "red", "Predčasné starobné": "brown", "Invalidné": "blue", "Vdovské": "yellow", "Sirotské": "orange"})
ax.set(xlabel=xlabel, ylabel=ylabel);
```

```python
title='Jednotlivé výdavky Sociálnej poisťovne na dôchodky v období 2009-2020'
ylabel='v mld. €'
xlabel='Rok'


axes = data.plot.bar(figsize=(12,12), grid=True,
    rot=0, subplots=True, color={"Starobné": "red", "Predčasné starobné": "brown", "Invalidné": "blue", "Vdovské": "purple", "Sirotské": "orange"}
)
axes[1].legend(loc=2)
```
